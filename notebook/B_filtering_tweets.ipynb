```python
import requests
from newspaper import Article
from bs4 import BeautifulSoup
import pandas as pd
import time

# Function to extract article using newspaper3k
def extract_article_newspaper3k(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        formatted_content = f"Heading:\n{article.title}\n\nArticle content:\n{article.text}"
        return formatted_content, 'newspaper3k'
    except Exception as e:
        return None, 'Failed'

# Function to extract article using BeautifulSoup
def extract_article_beautifulsoup(url):
    try:
        response = requests.get(url, timeout=30)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')
        heading = soup.find('h1')
        heading_text = heading.get_text(strip=True) if heading else 'No Heading Found'
        article_body = soup.find_all('p')
        article_content = "\n".join([p.get_text(strip=True) for p in article_body if p.get_text(strip=True)])
        formatted_content = f"Heading:\n{heading_text}\n\nArticle content:\n{article_content}"
        return formatted_content, 'BeautifulSoup'
    except Exception as e:
        return None, 'Failed'

# Function to attempt extraction first with newspaper3k, then BeautifulSoup
def extract_article_content(url):
    content, method = extract_article_newspaper3k(url)
    if method == 'Failed':
        content, method = extract_article_beautifulsoup(url)
    return content, method

# Create new columns to store extracted article content and method used
df_political['extracted_article_content'] = None
df_political['extraction_method'] = None

# Iterate over the rows and extract content from each URL
for index, row in df_political.iterrows():
    url = row['team_c2_extracted_url']
    if pd.notna(url):
        content, method = extract_article_content(url)
        df_political.at[index, 'extracted_article_content'] = content
        df_political.at[index, 'extraction_method'] = method

    # Save the DataFrame every 5 rows
    if (index + 1) % 5 == 0:
        df_political.to_csv('/content/drive/MyDrive/swati_team_c2/political_tweets_partial.csv', index=False)
        print(f"Saved partial DataFrame at index {index}")

    # Introduce a delay between requests
    time.sleep(5)

df_political.to_csv('/content/drive/MyDrive/swati_team_c2/political_tweets.csv', index=False)
df_political.head()
```

# Dividing dataset for manual annotations

```python
import pandas as pd
import csv

df_pol = pd.read_csv('/content/drive/MyDrive/swati_team_c2/political_tweets_partial.csv')
df_pol = df_pol[df_pol['extraction_method'] != 'Failed']
df_pol = df_pol.reset_index(drop=True)

df_pol = df_pol[df_pol['extracted_article_content'].notna() & (df_pol['extracted_article_content'] != '')]

# Further filter to remove rows with placeholder text in 'extracted_article_content'
placeholder_text = "Heading:\nx.com\n\nArticle content:\n"
df_pol = df_pol[df_pol['extracted_article_content'] != placeholder_text]

# Additionally, remove rows where 'team_c2_extracted_url' points to a video URL
# Assuming video URLs have certain keywords, e.g., "youtube.com" or "vimeo.com", add more keywords if necessary
video_keywords = ['youtube.com', 'vimeo.com', 'dailymotion.com']
df_pol = df_pol[~df_pol['team_c2_extracted_url'].str.contains('|'.join(video_keywords), case=False, na=False)]
df_pol = df_pol.reset_index(drop=True)
num_unique_cases = 150
num_overlapping_cases = 50
total_required_rows = num_unique_cases * 4 + num_overlapping_cases * 2  # 600 rows

if len(df_pol) < total_required_rows:
    raise ValueError("Not enough rows with valid 'extracted_article_content' for this distribution.")

unique_cases = df_pol.iloc[:num_unique_cases * 4].reset_index(drop=True)
overlapping_cases = df_pol.iloc[num_unique_cases * 4: total_required_rows].reset_index(drop=True)

# Distribute data to annotators
annotator_1_df = pd.concat([
    unique_cases.iloc[0:num_unique_cases].reset_index(drop=True),
    overlapping_cases.iloc[0:num_overlapping_cases].reset_index(drop=True)
], ignore_index=True)

annotator_2_df = pd.concat([
    unique_cases.iloc[num_unique_cases:num_unique_cases * 2].reset_index(drop=True),
    overlapping_cases.iloc[0:num_overlapping_cases].reset_index(drop=True)
], ignore_index=True)

annotator_3_df = pd.concat([
    unique_cases.iloc[num_unique_cases * 2:num_unique_cases * 3].reset_index(drop=True),
    overlapping_cases.iloc[num_overlapping_cases:num_overlapping_cases * 2].reset_index(drop=True)
], ignore_index=True)

annotator_4_df = pd.concat([
    unique_cases.iloc[num_unique_cases * 3:num_unique_cases * 4].reset_index(drop=True),
    overlapping_cases.iloc[num_overlapping_cases:num_overlapping_cases * 2].reset_index(drop=True)
], ignore_index=True)

annotator_1_df.to_excel('annotator_1.xlsx', index=False)
annotator_2_df.to_excel('annotator_2.xlsx', index=False)
annotator_3_df.to_excel('annotator_3.xlsx', index=False)
annotator_4_df.to_excel('annotator_4.xlsx', index=False)
print("Data distributed across annotators with 200 cases each, after filtering out invalid content, and saved to Excel files.")
```

```python
def extract_article_content(url):
    try:
        response = requests.get(url, timeout=30)  # Set a timeout to avoid hanging requests
        response.raise_for_status()  # Raise an exception if the request was unsuccessful
        soup = BeautifulSoup(response.content, 'html.parser')
        heading = soup.find('h1')  # Assuming article titles are in <h1> tags
        heading_text = heading.get_text(strip=True) if heading else 'No Heading Found'
        article_body = soup.find_all('p')
        article_content = "\n".join([p.get_text(strip=True) for p in article_body if p.get_text(strip=True)])
        formatted_content = f"Heading:\n{heading_text}\n\nArticle content:\n{article_content}"
        print("article extracted!",flush=True)
        return formatted_content
    except Exception as e:
        return f"Failed to extract content from {url}. Error: {e}"

def extract_articles_in_parallel(urls):
    with ThreadPoolExecutor(max_workers=10) as executor:  # Set an appropriate number of workers
        results = list(executor.map(extract_article_content, urls))
    return results

# Apply this parallel processing to the DataFrame
urls = df_political['team_c2_extracted_url'].dropna().tolist()  # Extract valid URLs
extracted_contents = extract_articles_in_parallel(urls)

# Add the extracted content back into the DataFrame
df_political.loc[df_political['team_c2_extracted_url'].notna(), 'extracted_article_content'] = extracted_contents
```
